```python
import requests
import pandas as pd 
import json
import yaml
```

```python
df_squawk7700 = pd.read_csv("../data/raw/squawk7700_metadata.csv")
registrations = df_squawk7700['registration']
```

```python
# YAML Credentials for API Keys
conf = yaml.load(open('../login/credentials/aerodatabox_api.yml'),Loader=yaml.SafeLoader)
api_key_rene = conf['rene']['api_key']
api_key_ralf = conf['ralf']['api_key']
api_list = [api_key_rene,api_key_ralf]

# Query parameters
params = {
    "withImage": "false",
    "withRegistrations": "true"
}

# Empty lists for responses and failed responses
reg_response_list = []
failed_reg_response_list = []

for i, reg in enumerate(registrations) :
    # Header for API
    headers = {
        "accept": "application/json",
        "x-magicapi-key": api_list[i%2]
    }

    # Define the search parameters per API documentation
    search_by = 'Reg' 
    search_param =  reg 
    # API endpoint
    url = f"https://api.magicapi.dev/api/v1/aedbx/aerodatabox/aircrafts/{search_by}/{search_param}"

    # Send the GET request
    response = requests.get(url, headers=headers, params=params)

    # Check the response
    if response.status_code == 200:
        # print("Response:", response.json())  # Parsed JSON response
        print(response.json())
        print(f"Ralf, I love you {i} out of {len(registrations)} <3<3<3<3<3<3<3<3<3")
        reg_response_list.append(response.json())
    else:
        print(f"Failure at iteration {i}, ICAO24 code: {registrations}")
        print(f"Error {response.status_code}: {response.text}")
        failed_reg_response_list.append(reg)
        continue

# Filepath to save the list
file_path_success = "../data/raw/holy_grail_registration_responses.json"

# Save the list to a JSON file
with open(file_path_success, "w") as file:
    json.dump(reg_response_list, file)
    print(f"List saved to {file_path_success}")

# Filepath to save the list
file_path_failure = "../data/raw/shit_grail_registration_responses.json"

# Save the list to a JSON file
with open(file_path_failure, "w") as file:
    json.dump(failed_reg_response_list, file)
    print(f"List saved to {file_path_failure}")
```
