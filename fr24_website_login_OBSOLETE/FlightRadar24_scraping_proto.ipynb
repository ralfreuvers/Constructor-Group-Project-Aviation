```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC

import time
import yaml
import pandas as pd 
```

```python

# YAML Credentials
conf = yaml.load(open('loginDetails.yml'),Loader=yaml.SafeLoader)
fr24_email = conf['fr24_user']['email']
fr24_password = conf['fr24_user']['password']

# Path to your ChromeDriver
chromedriver_path = "./website_login/chromedriver"  # Replace with your actual path

# Initialize WebDriver
driver = webdriver.Chrome()

try:
    # Open the login page
    driver.get("https://www.flightradar24.com/user/signin")

    # Wait for the page to load and locate the email and password fields
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.NAME, "email")))

    # Crawling Courtesy
    time.sleep(1)

    # Enter your login details
    email_field = driver.find_element(By.NAME, "email")
    password_field = driver.find_element(By.NAME, "password")
    login_button = driver.find_element(By.CSS_SELECTOR, "button[type='submit']")

    email_field.send_keys(fr24_email)
    password_field.send_keys(fr24_password)

    # Click the login button
    login_button.click()

    # Crawling Courtesy
    time.sleep(1)

    print("Login successful!")

except Exception as e:
    print(f"An error occurred at login: {e}")
```
